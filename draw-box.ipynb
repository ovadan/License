```python
import pandas as pd
from PIL import Image
import cv2
```

```python
full_labels = pd.read_csv("output.csv",header=None);
full_labels.columns = [['filename','plate_number','xmin', 'ymin', 'xmax', 'ymax','width','height']]
#full_labels.to_csv("output2.csv", encoding='utf-8')
full_labels.head()
full_labels.to_csv('example.csv')
```

```python
def draw_boxes(image_name):
    selected_value = full_labels[full_labels.filename == image_name]
    img = cv2.imread('generated_images/{}'.format(image_name))
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
Image.fromarray(draw_boxes('1.png'))
```
